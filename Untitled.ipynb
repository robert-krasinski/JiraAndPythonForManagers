### Stable velocity

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
stable = pd.read_csv('StableVelocity.csv', delimiter=';')

stable['velocityAvg'] =  round(stable['velocity'].shift().rolling(3, min_periods=1).mean(), 2)
stable['error'] = round(stable['velocity'] - stable['velocityAvg'], 2)
# rollAvg['estVelocityWBugs'] = round(rollAvg['capacity'] * rollAvg['ratioMean3'] + 2.64, 3) 
# rollAvg['error'] = rollAvg['estVelocityWBugs'] - rollAvg['velocityWBugs']
#stable

```

```python
_ = plt.xticks( stable.index.values, stable.sprintNo ) # location, labels
_ = plt.plot( stable['velocity'])
_ = plt.plot( stable['velocityAvg'])

_ = plt.xlabel('Sprint')
_ = plt.ylabel('velocity')
_ = plt.legend(['velocity', 'rolling average'])
_ = plt.show()
```

### Growing velocity

## Residuals analysis and prediction intervals
## Improvement on examples

```python

errors = stable.error.fillna(0)
_ = plt.hist(errors, bins=5, density=True, alpha=0.6, color='g')

# # Fit a normal distribution to the data:
mu, std = norm.fit(errors)


# Plot the PDF.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)
title = "Residue rollAvg Fit results: mu = %.2f,  std = %.2f" % (mu, std)
plt.title(title)

_ = plt.show()
```
